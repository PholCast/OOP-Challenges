```python
from abc import ABC, abstractmethod, abstractproperty
class Employee(ABC):
  @property
  @abstractmethod
  def first_namep(self):
    pass

  @property
  @abstractmethod
  def last_namep(self):
    pass

  @abstractmethod
  def full_name()-> str:
    pass
  

  @abstractmethod
  def get_salary():
    pass


class FullTimeEmployee(Employee):
  def __init__(self,first_name, last_name,salary):
    self.first_name = first_name
    self.last_name = last_name
    self.salary = salary
  
  @property
  def first_namep(self):
    return self.first_name

  @first_namep.setter
  def first_namep(self,first_name_value):
    self.first_name = first_name_value

  @property
  def last_namep(self):
    return self.last_name

  @last_namep.setter
  def last_namep(self,last_name_value):
    self.last_name = last_name_value
      
  
  def get_salary(self):
    return f"Salary: ${self.salary}"
  
  def full_name(self):
    return f"{self.first_name} {self.last_name}, a FulltimeEmployee"


class HourlyEmployee(Employee):
    def __init__(self,first_name, last_name,worked_hours,rate):
      self.first_name = first_name
      self.last_name = last_name
      self.worked_hours = worked_hours
      self.rate = rate

    @property
    def first_namep(self):
      return self.first_name

    @first_namep.setter
    def first_namep(self,first_name_value):
      self.first_name = first_name_value

    @property
    def last_namep(self):
      return self.last_name

    @last_namep.setter
    def last_namep(self,last_name_value):
      self.last_name = last_name_value

    @property
    def worked_hoursp(self):
      return self.worked_hours
    
    @worked_hoursp.setter
    def worked_hoursp(self,hours_value):
      self.worked_hours = hours_value
    
    @property
    def ratep(self):
      return self.rate
    
    @ratep.setter
    def ratep(self,rate_value):
      self.rate = rate_value

    def full_name(self):
      return f"{self.first_name} {self.last_name}, a hourlyEmployee"
    
    def get_salary(self):
      return f"Salary: ${self.rate * self.worked_hours}"

def function(array):
  for i in array:
    print(i.full_name())
    print(i.get_salary())
    print("\n")

full_1 = FullTimeEmployee("Phol","Castaneda",850300)
full_2= FullTimeEmployee("Esteban", "Gallego",940000)

hourly_1 = HourlyEmployee("Neithan","Gomez",6,33000) 
hourly_2 = HourlyEmployee("Daniela","Higuita",7,36000) 
arr = [full_1,hourly_2,full_2,hourly_1]

function(arr)
```

```python
from abc import ABC, abstractmethod, abstractproperty

class Animal(ABC):

  @property
  @abstractmethod
  def legs(self) -> int:
    pass

  @abstractmethod
  def walk(self):
    pass
  
  @abstractmethod
  def eat(self):
    pass


class Pet(ABC):
  @property
  @abstractmethod
  def name():
    pass

  @name.setter
  @abstractmethod
  def name():
    pass

  @abstractmethod
  def play():
    pass



class Spider(Animal):
  def __init__(self,legs):
    self.__legs = legs

  @property
  def legs(self) -> int:
    return self.__legs
  
  @legs.setter
  def legs(self,legs_value):
    self.__legs = legs_value
  
  def eat(self):
    print("The Spider is eating...")
  
  def walk(self):
    print("Spider walking")

class Cat(Animal,Pet):
  def __init__(self,legs,name):
    self.__legs = legs
    self.__name=name

  @property
  def legs(self) -> int:
    return self.__legs
  
  @legs.setter
  def legs(self,legs_value):
    self.__legs = legs_value
  
  def eat(self):
    print("The Cat is eating...")
  
  def walk(self):
    print("Cat walking...")
  
  @property
  def name(self):
    return self.__name

  @name.setter
  def name(self,name_value):
    self.__name = name_value
  
  def play(self):
    print(f"{self.name} is playing")



class Fish(Animal,Pet):
  def __init__(self,legs,name):
    self.__legs = legs
    self.name=name

  @property
  def legs(self) -> int:
    return self.__legs
  
  @legs.setter
  def legs(self,legs_value):
    self.__legs = legs_value
  
  def eat(self):
    print("The Fish is eating...")
  
  def walk(self):
    print("Fish swimming...")
  
  @property
  def name(self):
    return self.name

  @name.setter
  def name(self,name_value):
    self.name = name_value
  
  def play(self):
    print(f"{self.name} is playing")



s=Spider(5)
print(s.legs)
s.legs=7
print(s.legs)
s.eat()
s.walk()
print("\n"*2)


cat=Cat(4,"manolo")

print(cat.legs)
cat.eat()
cat.walk()
cat.play()
cat.name="wilson"
print(cat.name)




```
